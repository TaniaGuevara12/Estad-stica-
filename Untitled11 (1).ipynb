# ***Introducción***

En la actualidad, los trastornos relacionados con la salud mental, como la ansiedad y la depresión, representan una creciente preocupación a nivel mundial. Diversos factores pueden influir en el desarrollo de estos padecimientos, siendo la edad uno de los aspectos que más atención ha recibido en estudios recientes. El presente trabajo tiene como objetivo analizar la relación entre la edad y los niveles de ansiedad y depresión mediante herramientas estadísticas aplicadas a un conjunto de datos reales.

# ***Hipótesis***

Hipótesis nula (H₀): No existe una diferencia significativa en los niveles de ansiedad entre los distintos grupos de edad.

Hipótesis alternativa (H₁): Existen diferencias significativas en los niveles de ansiedad entre los distintos grupos de edad.

# ***Marco teórico***

La ansiedad y la depresión son trastornos mentales comunes que pueden afectar a personas de todas las edades. Sin embargo, estudios previos han demostrado que la prevalencia y la intensidad de estos trastornos pueden variar a lo largo de la vida. La adolescencia, la adultez temprana y la vejez son etapas especialmente críticas. A través del análisis estadístico, es posible determinar si existen diferencias significativas en los niveles de ansiedad según los grupos de edad, lo cual puede ayudar a focalizar estrategias de prevención y tratamiento.

# ***Metodología***



*   Se importaron los datos desde un archivo CSV alojado en GitHub.

*   Se eliminaron todas las filas con valores faltantes (NaN) usando dropna().

*   Se revisaron las columnas y los valores únicos en la variable Age para entender la estructura del dataset.

*   Se utilizó un modelo de regresión lineal (ols) Esto indica que se evalúa si existen diferencias significativas en el puntaje de ansiedad según los grupos de edad.



*   Se aplicó un ANOVA (Análisis de Varianza) al modelo para comprobar la significancia de la variable Age.
*   Al no encontrarse diferencias significativas, se eliminó la columna Age del dataset para futuras exploraciones.

# ***Resultados***


*   El modelo de regresión lineal se ejecutó correctamente con la variable "Anxiety_Score" como dependiente y "Age" como categórica.
*   El análisis ANOVA indicó que:

No existen diferencias estadísticamente significativas en los niveles de ansiedad según el grupo de edad.

Esto se refleja en un valor de p (p-value) mayor a 0.05, lo que lleva a no rechazar la hipótesis nula.


*   Por lo tanto, la edad no se considera un factor influyente en la ansiedad en este conjunto específico de datos.
*   Finalmente, se eliminó la variable Age para concentrar el análisis en otras variables potencialmente más relevantes.














# ***Conclusión***
Si bien la edad ha sido tradicionalmente considerada un factor importante en el desarrollo de trastornos de ansiedad, los resultados de este análisis específico no evidencian una relación significativa entre la edad y los niveles de ansiedad. Esto no invalida la importancia del factor edad, sino que pone de manifiesto la necesidad de considerar otras variables contextuales y demográficas para un análisis más completo. Se recomienda continuar con estudios más amplios y multivariados que integren otras dimensiones como género, nivel socioeconómico, y antecedentes familiares.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)

correlation_matrix = numeric_df.corr()
plt.figure(figsize=(15, 9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": 0.8})
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.title('Mapa de Calor de la Matriz de Correlación')
```

```python
#Selección de variables
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['Sleep_Hours', 'Social_Support_Score', 'Financial_Stress']]
Y = df['mayor_al_promedio']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, Y_train)

# Predecir
Y_pred = rf.predict(X_test)

# Evaluar
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

importancia = pd.Series(rf.feature_importances_, index=X.columns)
print(importancia.sort_values(ascending=False))
```

El modelo tiene un rendimiento bajo (accuracy 43%). Predice mejor la clase 1 (mayor recall), pero la precisión es similar en ambas clases. "Sleep_Hours" es la característica más importante (52.3%), seguida por "Social_Support_Score" (25.1%) y "Financial_Stress" (22.5%) para las predicciones del modelo Random Forest.

El modelo predice ambas clases con baja precisión (~43%). Sin embargo, para la clase 1 tiene un mejor recall (62%). La característica más importante para las predicciones es "Sleep_Hours" (52.3%), seguida de "Social_Support_Score" (25.1%) y "Financial_Stress" (22.5%).
