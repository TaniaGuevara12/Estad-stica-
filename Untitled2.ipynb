```python
# a) hipótesis de causalidad: Redacta una hipótesis sobre la causalidad entre las dos variables, a su vez, establece la variable dependiente y la variable independiente.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/TaniaGuevara12/Estad-stica-/refs/heads/main/mental_health.csv')
# Eliminar registros con valores faltantes
df
df.dropna(inplace=True)
df.head(60)

X = df["Year"] # Variab le independiente
Y = df["Eating disorders (share of population) - Sex: Both - Age: Age-standardized"] # Variable dependiente

# b)  recta de regresión ajustada: Realiza los cálculos necesarios para encontrar la recta de regresión ajustada. Incluye los procedimientos que te llevaron a la recta de regresión.
import matplotlib.pyplot as plt
plt.scatter(X, Y, color= 'purple')
plt.xlabel("Year")
plt.ylabel("Eating disorders (share of population) - Sex: Both - Age: Age-standardized")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()

# c) grafica los puntos y la recta de regresión
import statsmodels.api as sm
X_constante = sm.add_constant(X)
modelo = sm.OLS(Y, X_constante).fit()

b0, b1 = modelo.params

Fun = lambda x: b0 + b1 * x

Yc = Fun(X)

plt.plot(X, Yc, color= 'pink', linestyle = '--')

# d) Coeficiente de correlación y determinación: Realiza los cálculos necesarios para obtener la r de Pearson y el coeficiente de determinación. Incluye dichos cálculos
from scipy.stats import pearsonr
r,_ = pearsonr(X, Y)
print(f'coeficiente de correlacion: {r: 0.4f}\n')
print(f'coeficiente de determinacion: {r ** 2: 0.4f}\n')

# e)  Interpreta los resultados del coeficiente de correlación y el coeficiente de determinación.
# Coeficiente de correlación: El coeficiente de correlación de  0.0756  indica una correlación muy débil y positiva entre las variables estudiadas.
# Coeficiente de determinación: El coeficiente de determinación de   0.0057 es bajo Valor (cercano a 0) El modelo prácticamente no explica la relación entre las variables. 0.57% de explicación: La variable independiente tiene una influencia casi nula sobre la variable dependiente. Modelo poco útil: Si se usa para hacer predicciones, este modelo no es confiable, ya que la variable independiente apenas aporta información sobre la variabilidad de la dependiente.

# f) Calcula el intervalo de confianza del 95% para β_1 y β_0 ¿Qué dice el intervalo de confianza a la posibilidad de que β_1 sea igual a cero?
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}')

intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}')

# como el intervalo de confianza no incluye el valor 0 se puede rechazar la posibilidad de que b1=0.

# g) Realiza el gráfico de los residuales: A partir de este gráfico, menciona si los datos cumplen con los supuestos para la regresión: linealidad, normalidad en torno a la recta, homoscedasticidad (hay más, pero con estas nos bastan).
residuales = modelo.resid
plt.figure()
plt.scatter(X, residuales, color = 'brown')
plt.xlabel('Year')
plt.ylabel('Residuales')
plt.title('Gráfico de Residuales')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(0, color = 'pink', linestyle = '--')
plt.show()

from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f' valor_p Test de shapiro: {valor_p_sh: 0.4f}\n')

from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, X_constante)
print(f'Valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')

# Linealidad: En el gráfico de residuales, se observa una patrón curvo, lo que indica que no se cumple el supuesto de linealidad. Esto sugiere que la relación entre las variables no es estrictamente lineal y que el modelo de regresión lineal puede no ser el más adecuado. Se recomienda probar una transformación de variables o un modelo no lineal.
# Normalidad en torno a la recta: No se cumple, ya que el valor p de Shapiro es 0.0000 (menor a 0.05).
# Homoscedasticidad: No se cumple, pues el valor p de Breusch-Pagan es 0.0001, indicando varianza no constante.



























```
