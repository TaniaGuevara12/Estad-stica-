```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/TaniaGuevara12/Estad-stica-/refs/heads/main/data%20(2).csv')
df.head()

# a) Revisa si el dataset tiene valores nulos mediante df.info(), en dado caso, aplica df.dropna(inplace = True).
df.info()
df.dropna(inplace=True)

# b) Obten un mapa de calor con la matriz de correlacion. Determina las variables independientes y la variable dependiente y redacta una hipotesis.
import seaborn as sns
import matplotlib.pyplot as plt
correlacion_matrix = df.corr()
plt.figure(figsize=(15, 9))
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm', fmt='.2f',square=True,cbar_kws={'shrink':.82})
plt.title('Mapa de calor de la matrix de correlacion')

# c) Obten el modelo que permite relacionar las variables independientes y la variable dependiente.
X = df[['Duration','Maxpulse']]
Y = df['Calories']

import statsmodels.api as sm
X_constante = sm.add_constant(X)
model = sm.OLS(Y, X_constante).fit()
Yc = model.predict(X_constante)
plt.figure()
plt.scatter(Y,Yc, color = 'Purple')
print(model.summary())

# d) Obten el coeficiente de correlacion e interpreta el resultado.
from sklearn.metrics import r2_score
r2 = r2_score(Y,Yc)
print(f'El coeficiente de correlacion es {r2: 0.4}/n ')
residuales = model.resid

# e) Verifica los supuestos mediante un grafico "y" vs "y_prend", el test de Shapiro-Wilk y el test de Breusch-Pagan. Interpreta los resultados.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f'Valor p del test de Shapiro-Wilk es {valor_p_sh: 0.4f}')
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(model.resid, model.model.exog)
print(f'Valor p del test de Breusch-Pagan es {valor_p_bp: 0.4f}')
# Se rechaza Ho
# f) Tomando en cuenta la informacion obtenida, redacta una conclusion.
# Se rechaza Ho poruqe el valor es muy bajo.
```
