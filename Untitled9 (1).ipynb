```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['CreditScore',	'Age',	'Tenure',	'Balance']]
Y = df["EstimatedSalary"]

# Datos de entrenamiento y datos de prueba

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()

Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de correlacion: {r2: 0.4f}\n')
```

 Indica una relación prácticamente nula entre las dos variables analizadas. En términos estadísticos, esto significa que no hay una asociación lineal significativa entre ellas.


```python
b0, b1, b2, b3, b4 = modelo.params

Fun = lambda x1, x2, x3, x4: b0 + b1 * x1 + b2 * x2 + b3 * x3 + b4 * x4

# El salario estimado se puede calcular usando variables como el puntaje de crédito, edad, antigüedad laboral y saldo financiero , ya que reflejan estabilidad, experiencia y capacidad de pago.
# Estos datos permiten predecir ingresos mediante modelos estadísticos o de aprendizaje automático.

Fun(432, 60, 2, 120000)
```

df (Grados de Libertad)
Representa el número de valores que pueden variar libremente en el cálculo.

Cada variable independiente tiene 1 grado de libertad, lo que significa que se evalúan de manera individual.

Los residuos tienen 6377 grados de libertad, lo cual refleja el tamaño de la muestra menos el número de parámetros estimados.

✅ Conclusión: El tamaño de muestra es grande, pero la alta libertad de los residuos sugiere que las variables explicativas tienen un impacto limitado en la predicción del salario.

F (Estadístico F)
Indica la razón de variación explicada sobre la no explicada.

Valores muy bajos para CreditScore (0.000268) y Age (0.469869) sugieren un efecto débil sobre "EstimatedSalary".

Esto implica que el modelo no tiene un poder explicativo significativo.

✅ Conclusión: El modelo no es estadísticamente significativo, ya que los valores F bajos reflejan una variación explicada muy inferior a la variación residual.

PR(>F) (Valor p o Nivel de Significancia)
Mide la significancia estadística del efecto de cada variable.

Valores altos (> 0.05), como CreditScore (0.986941) y Age (0.493072), indican que ninguna variable es estadísticamente significativa en la predicción de "EstimatedSalary".

✅ Conclusión: No hay evidencia suficiente para afirmar que estas variables influyen significativamente en el salario estimado. El modelo tiene poca capacidad predictiva y no es útil para decisiones prácticas.

Conclusión General
El modelo de regresión múltiple utilizado presenta un poder explicativo muy limitado para predecir el salario estimado. Ninguna de las variables independientes es estadísticamente significativa, lo cual se refleja en los valores p elevados y el estadístico F bajo. Esto sugiere que otras variables más relevantes deberían incluirse para mejorar la predicción o, de lo contrario, considerar un modelo alternativo.
